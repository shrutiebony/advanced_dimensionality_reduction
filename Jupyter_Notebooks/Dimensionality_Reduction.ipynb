```
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
target_names = iris.target_names

# Step 2: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 3: Apply PCA for Dimensionality Reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_scaled)

# Create a DataFrame with the PCA results
pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Target'] = target

# Step 4: Visualization of the Reduced Dimensions
plt.figure(figsize=(10, 7))
for i, target_name in enumerate(target_names):
    plt.scatter(
        pca_df[pca_df['Target'] == i]['Principal Component 1'],
        pca_df[pca_df['Target'] == i]['Principal Component 2'],
        label=target_name
    )

plt.title('PCA on Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

# Step 5: Explained Variance
print("Explained Variance Ratio by Components:", pca.explained_variance_ratio_)
```

```
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
target_names = iris.target_names

# Create a DataFrame for the original data
original_df = pd.DataFrame(data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
original_df['Target'] = target

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply PCA for Dimensionality Reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_scaled)

# Create a DataFrame with the PCA results
pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Target'] = target

# Visualize Original Data (Sepal Length vs Sepal Width)
plt.figure(figsize=(16, 6))

# Plot Original Data
plt.subplot(1, 2, 1)
for i, target_name in enumerate(target_names):
    plt.scatter(
        original_df[original_df['Target'] == i]['Sepal Length'],
        original_df[original_df['Target'] == i]['Sepal Width'],
        label=target_name
    )
plt.title('Original Data: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()

# Plot PCA Transformed Data
plt.subplot(1, 2, 2)
for i, target_name in enumerate(target_names):
    plt.scatter(
        pca_df[pca_df['Target'] == i]['Principal Component 1'],
        pca_df[pca_df['Target'] == i]['Principal Component 2'],
        label=target_name
    )
plt.title('PCA Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()

```
